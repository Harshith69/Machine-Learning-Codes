```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#plotting auto correlation and partial auto correlation graphs for price datasets
#library for graph
import matplotlib as mpl 
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,6))
plot_acf(df1, lags=40, ax=ax[0])
ax[0].set_title("Autocorrelation Function (ACF)")

plot_pacf(df1, lags=40, ax=ax[1])
ax[1].set_title("Partial Autocorrelation Function (PACF)")

plt.tight_layout()
plt.show()
```

```python
# perform first-order differencing and plotting the graph
df1_diff = df1.diff(periods=1)

# plot the original time series
# plt.plot(df1)
# plt.title('Original Time Series')
# plt.show()

# plot the differenced time series
plt.plot(df1_diff)
plt.title('First order differencing of Time Series')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.show()
```

```python
#plotting auto correlation and partial auto correlation graphs after first order differentiation
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,6))
plot_acf(df1_diff, lags=40, ax=ax[0])
ax[0].set_title("Autocorrelation Function (ACF)")

plot_pacf(df1_diff, lags=40, ax=ax[1])
ax[1].set_title("Partial Autocorrelation Function (PACF)")

plt.tight_layout()
plt.show()
```

```python
#plotting the lag line chart to find correlation bw day lag period
import matplotlib.pyplot as plt
import pandas as pd

# Load data
# df = pd.read_csv("data.csv")

# Create lag line chart
plt.plot(df["Price"], df["Price"].shift(1), "o", alpha=0.5, color='red')
plt.title("Lag Line Chart")
plt.xlabel("Price(t)")
plt.ylabel("Price(t+1)")
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
ax.set_facecolor('white')

# Load data and set index to datetime
df = pd.read_csv('/content/Wheat_data23.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Perform seasonal decomposition
decomposition = sm.tsa.seasonal_decompose(df, model='additive', period=12)

# Plot trend and residuals
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(decomposition.trend, color='indigo', label='Trend')
ax.plot(decomposition.resid, color='red', alpha=0.5, label='Residuals')
ax.legend(loc='upper left')
plt.show()

```
